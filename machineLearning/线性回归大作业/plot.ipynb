```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']=['SimHei'] #显示中文标签
plt.rcParams['axes.unicode_minus']=False
```

```python
df = pd.read_excel('一层隐藏成神经元.xlsx')
nerou = df.iloc[:, 0]
train = df.iloc[:, 1]
test = df.iloc[:, 2]
```

```python
plt.plot(nerou, train, label='训练集')
plt.plot(nerou, test, label='验证集')
plt.ylabel('Loss')
plt.xlabel('隐藏层单元个数')
plt.grid()
plt.legend()
plt.savefig('../../期末作业/figures/两层隐藏层单元个数图.png', dpi=900)
```
