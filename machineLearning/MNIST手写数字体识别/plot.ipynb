```python
df1 = pd.read_csv('../../期末作业/手写数字体识别临时文件/MLP学习率1.csv')
df2 = pd.read_csv('../../期末作业/手写数字体识别临时文件/MLP学习率2.csv')
df3 = pd.read_csv('../../期末作业/手写数字体识别临时文件/MLP学习率3.csv')

neure = df1.iloc[:, 0]
error_1 = df1.iloc[:, 3]
error_2 = df2.iloc[:, 3]
error_3 = df3.iloc[:, 3]
```

```python
plt.figure(dpi=144)
plt.plot(neure, error_1, 'ro')
plt.plot(neure, error_1, 'r', label='一层隐藏层')
plt.plot(neure, error_2, 'g^')
plt.plot(neure, error_2, 'g', label='两层隐藏层')
plt.plot(neure, error_3, 'bs')
plt.plot(neure, error_3, 'b', label='三层隐藏层')
plt.xlabel('学习率')
plt.ylabel('测试集上预测错误数')
# plt.xticks(neure)
plt.legend()
# plt.savefig('../../期末作业/figures/三个层隐藏层学习率变化图 .png', dpi=900)
```
